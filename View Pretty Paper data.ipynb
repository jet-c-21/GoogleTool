```python
from pprint import pp
import pandas as pd
from google_tool import init_resource_service
from google_tool.drive import FolderRetriever
```

```python
CLIENT_SECRET_FILE = 'jcv_oauth.json'
API_NAME = 'drive'
API_VERSION = 'v3'
SCOPES = ['https://www.googleapis.com/auth/drive']
```

```python
service = init_resource_service(CLIENT_SECRET_FILE, API_NAME, API_VERSION, SCOPES)
```

```python
fr = FolderRetriever(service)
```

```python
folder_id = '1uR5XCZ3zSnbiuTHCB9uez9JHXUdndm-I'  # PrettyPaper/data
```

```python
%%time
folder_df = fr.get_all_folder(folder_id)
```

```python
folder_df
```

```python
folder_df = folder_df[folder_df['level']==2]
```

```python
folder_df = folder_df[['path', 'id']]
folder_df['mid'] = folder_df['path'].apply(lambda x: int(x.split('/')[1].split('-')[0]))
folder_df['subject'] = folder_df['path'].apply(lambda x: x.split('/')[1].split('-')[1])
folder_df = folder_df.sort_values(['mid', 'path'], ascending = [True, True]).reset_index(drop=True)
folder_df
```
